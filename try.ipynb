```python
import numpy as np

# Portfolio weights
weights = np.array([0.5, 0.5])  # 50% in each asset

# Expected returns
returns = np.array([0.1, 0.08])  # 10% and 8%

# Covariance matrix (volatility and correlation combined)
cov_matrix = np.array([[0.04, 0.006], [0.006, 0.03]])

# Expected portfolio return
expected_return = np.dot(weights, returns)

# Portfolio variance
portfolio_variance = np.dot(weights.T, np.dot(cov_matrix, weights))

# Portfolio standard deviation (risk)
portfolio_std = np.sqrt(portfolio_variance)

print(f"Expected Return: {expected_return:.2%}")
print(f"Portfolio Risk (Std Dev): {portfolio_std:.2%}")
```

```python
import numpy as np

P0 = 100
Pn = 121
n = 2  # years

# Simple Return
simple_return = (Pn - P0) / P0

# Log Return
log_return = np.log(Pn / P0)

# Annualized Compound Return
compound_return = (Pn / P0)**(1/n) - 1

print(f"Simple Return: {simple_return:.2%}")
print(f"Log Return: {log_return:.2%}")
print(f"Annualized Compound Return: {compound_return:.2%}")
```
